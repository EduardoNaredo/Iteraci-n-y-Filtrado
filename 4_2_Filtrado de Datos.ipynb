```python
#Importamos librerías requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('listings.csv')
df
```

```python
#Los registros de 5 host names diferentes
hosts = df.loc[[0, 1, 2, 3, 4]]
hosts
```

```python
#Convertir archivo filtrado a CSV
hosts.to_csv("hosts.csv")
```

```python
#Los registros de los host que se hallan unido a Airbnb “despues del año 2020”
df["host_since"] = pd.to_datetime(df["host_since"])

hostsince = df[df["host_since"] > "2020-12-31"]
hostsince
```

```python
#Convertir archivo filtrado a CSV
hostsince.to_csv("hostsince.csv")
```

```python
pd.options.display.max_columns = None
```

```python
#Los registros de los host que responden “a más tardar en 1 día”
responsetime = df[~df["host_response_time"].isin(['a few days or more'])& df["host_response_time"].notna()]
responsetime
```

```python
#Convertir archivo filtrado a CSV
responsetime.to_csv("responsetime.csv")
```

```python
#Los registros de las filas “Múltiplos de 200… (400, 600, 800, etc.)”
multiplos200 = df.iloc[0::200]
multiplos200
```

```python
#Convertir archivo filtrado a CSV
multiplos200.to_csv("multiplos200.csv")
```

```python
#Los registros de los tipos de cuarto: “Entire home” y “Private room”
cuartos = df[df["room_type"].isin(['Entire home/apt', 'Private room'])]
cuartos
```

```python
#Convertir archivo filtrado a CSV
cuartos.to_csv("cuartos.csv")
```

```python
#Los registros que “a lo mucho cuenten con 4 camas”
camas4=df[df["beds"] < 5]
camas4
```

```python
#Convertir archivo filtrado a CSV
camas4.to_csv("camas4.csv")
```

```python
#Los registros que “sean superhost”
superhost = df[df["host_is_superhost"] == "t"]
superhost
```

```python
#Convertir archivo filtrado a CSV
superhost.to_csv("superhost.csv")
```

```python
#Los registros de las columnas “impares”
impares = df.iloc[:, 1::2]
impares
```

```python
#Convertir archivo filtrado a CSV
impares.to_csv("impares.csv")
```
